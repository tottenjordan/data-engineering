### **Factorial Function using Recursion**

```
def fact(x):
  if x == 1:
    return 1
  else:
    return x * fact(x-1)
```

# **Executive Summary**

**The plot "Problem Size vs Execution Time" shows the computational difference between using For-Loops and recursive functions. The recursive function uses the call stack to save iterations while the for loop does not. The for loop just loops through the iteration of calculating the factorial of every required number. The stack calls used by the recursive function allow it to calculate quicker than the for loop. The trade-off is the recursive function's use of memory. As the computational requirements of the recrusive function grow, the call stack grows. Overloading the stack could result in using all the computers memory. **
